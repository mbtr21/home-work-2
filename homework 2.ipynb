```julia
function heurestic1(start,stop)
    i = stop[1] - start[1]
    j = stop[2] - start[2]
    result = round(sqrt(i^2 + j^2))
return result 
end
function heurestic2(start,stop)
    i = stop[1] - start[1]
    j = stop[2] - start[2]
    result = ((i^2 + j^2))
return result 
end
```

```julia

function A_star(Area,start,stop,heurestic)
    g = 0
    frontier = []
    visited = []
    x1 = [[start] ,0,heurestic(start,stop)]
    push!(visited ,start)
    push!(frontier , x1)
    final_path = []
    while size(frontier)[1]!=0
#println("frontier is ",frontier)
#println("visited is ",visited)
    A = popfirst!(frontier)
    location = last(A[1])
    #rintln("path is",A[1])
    if location == stop 
        final_path = copy(A[1])
        return (final_path)
        break 
    else
        if (location[1] + 1<= size(Area)[1])  
            if Area[location[1]+1,location[2]] != 1
                 if [location[1]+1,location[2]] in visited
                    g =1
                else
                   
                    push!(visited,[location[1]+1,location[2]])
                    cost = A[2] + 2
                    path = copy(A[1])
                    push!(path,[location[1]+1,location[2]])
                    new = []
                    push!(new,path)
                    push!(new,cost)
                    push!(new,heurestic1([location[1]+1,location[2]],stop))
                    push!(frontier,new)
               
        end
        end
    end
     if (location[1] - 1 >= 1)      
        if Area[location[1]-1,location[2]]!=1
                 if [location[1]-1,location[2]] in visited
                g = 1
                else
                    push!(visited,[location[1]-1,location[2]])
                    cost = A[2] + 3
                    path = copy(A[1])
                    push!(path,[location[1]-1,location[2]])
                    new = []
                    push!(new,path)
                    push!(new,cost)
                    push!(new,heurestic1([location[1]-1,location[2]],stop))
                    push!(frontier,new)
        end  
        end
    end
     if(location[2] - 1 >= 1) 
        if (Area[location[1],location[2] - 1] != 1)
                 if [location[1],location[2]-1] in visited
                    g = 1
                else
                    
                    push!(visited,[location[1],location[2]-1])
                    cost = A[2] + 1
                    path = copy(A[1])
                    push!(path,[location[1],location[2] - 1])
                    new = []
                    push!(new,path)
                    push!(new,cost)
                    push!(new,heurestic1([location[1],location[2]-1],stop))
                    push!(frontier,new)
        end
        end
    end
#println("frontier is ",frontier)
     if (location[2] +1 <= size(Area)[2]) 
        if (Area[location[1],location[2] + 1] != 1)
                 if [location[1],location[2]+1] in visited
                    g = 1
                else
                    
                    push!(visited,[location[1],location[2]+1])
                    cost = A[2] + 1
                    path = copy(A[1])
                    push!(path,[location[1],location[2] + 1])
                    new = []
                    push!(new,path)
                    push!(new,cost)
                    push!(new,heurestic1([location[1],location[2]+1],stop))
                    push!(frontier,new)
        
    end      
    end
    end
sort!(frontier , by = x ->x[2] + x[3] ) 
end
end
end
list = []
for i in 1:7
    t = map(rand, [100:500, 100:500])
    sort!(t)
    push!(list,t)
end


            
```
